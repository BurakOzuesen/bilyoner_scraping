```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('match_odds_wide.csv')
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=1)
```

```python
test_row = test_df.iloc[0]
test_row
```

```python
test_row["homeTeam"], test_row["awayTeam"]
```

```python
# İşlem yapılmaması gereken ilk 6 sütun
meta_cols = ['match_date', 'match_time', 'tournament', 'match_id', 'homeTeam', 'awayTeam', 'firstHalfHomeGoal','firstHalfAwayGoal','totalHomeGoal','totalAwayGoal','homeCorner','awayCorner']
numeric_cols = [col for col in train_df.columns if col not in meta_cols]
test_row = test_df.iloc[0]


# Distance'ları tutacak liste
distances = []

for idx, train_row in train_df.iterrows():
    train_values = train_row[numeric_cols]
    test_values = test_row[numeric_cols]

    common_mask = ~(train_values.isna() | test_values.isna())

    if common_mask.sum() == 0:
        dist = np.inf
    else:
        diff_squared = (train_values[common_mask] - test_values[common_mask]) ** 2
        dist = np.sqrt(diff_squared.sum())

    distances.append((idx, dist))

# Distance sonuçları
distance_df = pd.DataFrame(distances, columns=['index', 'distance'])

# Train df'den ilk 6 sütunu ve index'i al
meta_df = train_df[meta_cols].copy()
meta_df = meta_df.reset_index().rename(columns={'index': 'train_index'})

# Distance ile birleştir
merged_df = pd.merge(meta_df, distance_df, left_on='train_index', right_on='index')

# Distance'a göre sırala
sorted_result = merged_df.sort_values(by='distance').reset_index(drop=True)

# Sonuç
print(sorted_result)
```

```python
sorted_result.to_csv('sorted_distances.csv', index=False)
```
