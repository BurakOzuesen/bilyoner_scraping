```python
with open('yorumlu_maclar.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()
    print("Toplam satır sayısı:", len(lines))
```

```python
lines
```

```python
preds = []
for line in lines:
    orginal_line = line
    line = line.split()
    max_count = 0
    for word in line:
        try:
            word = int(word)
            if word > max_count:
                max_count = word
        except ValueError:
            continue
    preds.append([max_count, orginal_line])
```

```python
preds.sort(reverse=True)
preds
```
